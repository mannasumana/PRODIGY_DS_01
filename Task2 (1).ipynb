```python

fig, axes = plt.subplots(2, 4, figsize=(16, 10))
sns.countplot(x='Survived', data=df, ax=axes[0, 0])
sns.countplot(x='Pclass', data=df, ax=axes[0, 1])
sns.countplot(x='Sex', data=df, ax=axes[0, 2])
sns.countplot(x='SibSp', data=df, ax=axes[0, 3])
sns.countplot(x='Parch', data=df, ax=axes[1, 0])
sns.countplot(x='Embarked', data=df, ax=axes[1, 1])

# Distribution plots
sns.histplot(df['Fare'], kde=True, ax=axes[1, 2])
sns.histplot(df['Age'].dropna(), kde=True, ax=axes[1, 3])

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
figbi, axesbi = plt.subplots(2, 4, figsize=(16, 10))
df.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
df.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
df.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
df.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
df.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=df,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=df,ax=axesbi[1,2])
```

```python
group = df.groupby(['Pclass', 'Survived']) 
pclass_survived = group.size().unstack() 
  
# Heatmap - Color encoded 2D representation of data. 
sns.heatmap(pclass_survived, annot = True, fmt='d')
```

```python

# Assuming 'df' is your DataFrame and it has the 'Parch', 'SibSp', and 'Survived' columns

# Create Family_Size column
df['Family_Size'] = df['Parch'] + df['SibSp']

# Adding a column for Alone (family size = 0 means Alone)
df['Alone'] = 0
df.loc[df.Family_Size == 0, 'Alone'] = 1

# Catplot for Family_Size vs Survived
sns.catplot(x='Family_Size', y='Survived', data=df, kind='point')  # Use 'point' for a clear comparison
plt.title('Survival vs Family Size')
plt.show()

# Catplot for Alone vs Survived
sns.catplot(x='Alone', y='Survived', data=df, kind='point')  # Use 'point' for a clear comparison
plt.title('Survival vs Alone')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

# Create crosstab of 'sex', 'Survived', and 'Pclass', with margins added
crosstab = pd.crosstab([df.sex, df.survived], df.pclass, margins=True)

# Apply a background gradient to the crosstab
styled_crosstab = crosstab.style.background_gradient(cmap='summer_r')


styled_crosstab
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df = sns.load_dataset("titanic")

# Create a figure with two subplots
f, ax = plt.subplots(1, 2, figsize=(20, 10))

# Histogram for passengers who did not survive (Survived = 0)
df[df['survived'] == 0].age.plot.hist(ax=ax[0], bins=20, edgecolor='black', color='red')
ax[0].set_title('Survived = 0')
ax[0].set_xticks(list(range(0, 85, 5)))  # Set x-ticks from 0 to 85 in steps of 5
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Frequency')

# Histogram for passengers who survived (Survived = 1)
df[df['survived'] == 1].age.plot.hist(ax=ax[1], bins=20, edgecolor='black', color='green')
ax[1].set_title('Survived = 1')
ax[1].set_xticks(list(range(0, 85, 5)))  # Set x-ticks from 0 to 85 in steps of 5
ax[1].set_xlabel('Age')
ax[1].set_ylabel('Frequency')


plt.show()
```
