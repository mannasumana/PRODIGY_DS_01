```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)  # For reproducibility
pizza_types = ['Margherita', 'Pepperoni', 'BBQ Chicken', 'Veggie', 'Hawaiian']
sizes = np.random.randint(8, 18, size=5)  
prices = np.random.uniform(5, 25, size=5)

df = pd.DataFrame({'Pizza Type': pizza_types, 'Size': sizes, 'Price': prices})




cols = ['Size', 'Price', 'Calories']  


for i in  df.columns.intersection(cols): 
        fig = plt.figure(figsize=(5, 5))
        plt.hist(df[i], color='blue', bins=5, edgecolor='black')
        plt.xlabel(i)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {i}')
        plt.show()
```
